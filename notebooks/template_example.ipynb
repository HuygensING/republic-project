# Example template in the 18th century 

As an example of a resolution template we take the formulaic appointment of new delegates. A typical resolution looks as follows (example 1757-03-31 resolution-6): "ONtfangen een Missive van de Heeren Gedeputeerde Staaten van de Provincie van Vriesland, geschreven te Leeuwaarden den een en twintighsten deeser loopende maand, houdende Creditif op den Heer Georg Sigismund van Hiemstra, als ordinaris Gecommitteerde ter Vergaderinge van haar Hoogh Mogende. WAAR op gedelibereert, en by haar Hoogh Mogende verklaart zynde, dat gemelden Heere wan Hiemstra haar aangenaam was, is den selven verwelkomt , en heeft dienvolgende sessie genoomen." 

(EN: "Receipt of a letter from the Lords of the Provincial States of Friesland, written in Leeuwarden on the twenty-first of this month, containing a letter of credit to Mr Georg Sigismund van Hiemstra, as ordinary member of the committee at the meeting of the Almighty. WHEREOF, having been duly deliberated upon, and having been declared by her Highness that the said Lord wan Hiemstra was agreeable to her, the same was welcomed, and took his session thereupon")

If we search the standard formulas of 1757 on the phrase 'heeft dienvolgende sessie' (EN:'thereupon his session'), this yields 85 formulas containing this formula, corresponding to 28 unique resolutions. This is 0.03 percent of the 250,506 formulaic expressions identified, or 0,1 percent of all 2919 resolutions in 1757.

The resolutions themselves is _very_ formulaic. A table output of a collatex comparison of the last section of a part of the resolutions shows there is a template underlying these appointment resolutions.

Even if they are not as frequent as some other types of resolutions, they are found throughout the eighteenth century. If we take the same formula as above, we find 2160 unique days on which the formulas appear. 

But searching for a single phrase shares the problem with many other text mining tasks, that it may both find too many and too few resolutions. The resolutions and the formulas contain _a lot_ of variations and are also subject to change over time.

In other words, we find the resolutions that confirm to a single formula, but it is by no means sure that the formula is an indication for the appointment template.

Because we do not have a way to determine templates yet, we therefore searched for different sub formulas in the template and compared the results. The queries were:

- 'heeft dienvolgende sessie', as treated above (characterized as 'sessie')
- '[Cc]reditif' (EN: 'credit letter') (characterized as 'creditif')
- 'verwell?e?komt' (EN: 'welcomes') (characterized as 'welkom')

This results in the following graph

```python
import pandas as pd
result = pd.read_csv('./csvs/appointment_templates.csv', sep='\t')
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

tick_spacing = 5
font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'style': 'italic',
        'size': 11,
        }

plt.rcParams.update({'font.size': 9, 'font.family': 'serif',})
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize=(10,15))
fig.tight_layout(pad=5.0)
result.set_index('year').plot.bar(ax=ax1)

ax1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax1.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax1.xaxis.set_minor_formatter(ticker.NullFormatter())
ax1.set_title("appointment formulas per year (1705-1796)", fontdict=font)
ax1.set_ylabel('templates')

result.set_index('year').sessie.plot.bar(ax=ax2, color='blue')
ax2.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax2.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax2.xaxis.set_minor_formatter(ticker.NullFormatter())
ax2.set_title("'sessie heeft genomen' per year (1705-1796)", fontdict=font)
ax2.set_ylabel('templates')

result.set_index('year').welkom.plot.bar(ax=ax3, color='orange')
ax3.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax3.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax3.xaxis.set_minor_formatter(ticker.NullFormatter())
ax3.set_title("'Creditif' per year (1705-1796)", fontdict=font)
ax3.set_ylabel('templates')

result.set_index('year').creditif.plot.bar(ax=ax4, color='green')
ax4.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax4.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax4.xaxis.set_minor_formatter(ticker.NullFormatter())
ax4.set_title("'verwelkomt' per year (1705-1796)", fontdict=font)
ax4.set_ylabel('templates')

plt.show()
```

The different formulas give roughly the same graph, but there are also significant differences, where one formula does appear in the resolution while another does not. As they are all part of a template, they often co-occur. A reliable search for these resolutions, however, depends on finding the template.

The variations in the graph call for further research, but we can already highlight two peaks. In 1748, when the Dutch Republic was under siege from France and there was a popular movement to install William IV of Orange as a stadtholder after a stadtholder-less period, resulted in a change of regents in many of the Dutch cities. This was reflected in the delegates the provinces sent to the States General as their representatives. In 1796, there was a government change in the spirit of the French revolution, provoked by a runover of the Dutch Republic by the French revolutionary army. This ended in the abolishment of the Republic as it had existed, but before that the delegates in the States General were replaced by new representatives on behalf of the revolutionary councils in the different provinces.



## Formulas to templates

The variation in the formulas makes it hard to bundle them into templates. Also, there are different combinations of formulas that make up different templates. There are also variations in the larger templates. For instance, when an request was approved by the States General, this was at times under certain provisions. At other times, the council gave additional orders for the execution. These can be seen as extensions of the basic template. Up front it seems unlikely they would follow a template pattern, but they may still contain some formulaic content.
