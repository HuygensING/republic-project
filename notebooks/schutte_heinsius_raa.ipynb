## Heinsius index

The correspondents is only a subset of the names in the Heinsius edition. The bulk is in the index

```python
heinsius_dir = "/Users/rikhoekstra/develop/namenindex/heinsius/out/"
```

```python
import os
heinsius_names = []
fls = [f for f in os.listdir(heinsius_dir) if '.xml' in f]
for f in fls:
    item = {}
    with open(os.path.join(heinsius_dir, f)) as infl:
        n = os.path.splitext(f)[0]
        item['nr'] = n
        ndoc = etree.parse(infl)
        p = [r for r in ndoc.getroot().getchildren() if r.tag == 'person'][0]
        persname = [r for r in p.getchildren() if r.tag == 'persName'][0]
        name = persname.text
        item['name'] = name
        b = [r for r in ndoc.getroot().getchildren() if r.tag == 'biography'][0]
        bc = b.getchildren()
        if len(bc)>0:
            text = bc[0].text
        else:
            text = ''
        item['text'] = text
        heinsius_names.append(item)
```

```python
import html
df_nms = pd.DataFrame.from_records(heinsius_names)
df_nms['text_unescaped'] = df_nms.text.apply(lambda x: html.unescape(x)) # we leave the unescaped text for possible searching
df_nms
```

```python
# this regex tries to get the living years from the description
import re
pat1 = re.compile("(?P<year1>(1+[0-9]{3})?)(?P<hyphen>[†-])?(?P<year2>1+[0-9]{3})+")

def match_years(x):
    res = pat1.search(x)
    if res:
        res = res.groupdict()
    return res
        
```

```python
#put the year in the data frame
pd.concat([df_nms, df_nms.text_unescaped.apply(lambda x: match_years(x)).apply(pd.Series)], axis=1)
```

```python
# this is from https://stackoverflow.com/questions/13636848/is-it-possible-to-do-fuzzy-match-merge-with-python-pandas
import difflib 

def fuzzy_merge(df1, df2, left_on, right_on, how='inner', cutoff=0.6):
    df_other= df2.copy()
    df_other[left_on] = [get_closest_match(x, df1[left_on], cutoff) 
                         for x in df_other[right_on]]
    return df1.merge(df_other, on=left_on, how=how)


def get_closest_match(x, other, cutoff):
    matches = difflib.get_close_matches(x, other, cutoff=cutoff, n=1)
    return matches[0] if matches else None

# def get_closest_match(x, other, cutoff):
#     matches = score_levenshtein_similarity_ratio(x, other)
#     return matches if matches else None
```
