```python
def fld2org(row):
    orglst = ['org_naam_wapen', 'org_naam_regiment',
       'org_naam_bataljon', 'org_naam_compagnie']
    clmns = [row[c] for c in row.keys() if c in orglst if not pd.isna(row[c])]
    result = ', '.join(clmns)
    return result
```

```python
import tarfile #import Tarfile
def compress_output_files(outputdir='.', outputfolder='out'):
    outputfile = os.path.join(outputdir, outputfolder)
    print('writing to', outputfile)
    tar = tarfile.open(f"{outputfile}.tar.gz", "w:gz")
    for name in os.listdir(outputfile):
        tar.add(outputfile + "/" + name, arcname=name)
    tar.close()
```
