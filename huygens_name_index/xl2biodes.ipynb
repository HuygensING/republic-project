```python
def sheet2person(row, name_publisher=''):
    gender_d={'M':1,'F':2}
    name = row['Person primary name in EMLO']
    variants = [v.strip() for v in row['Synonyms'].split(';')]
    gender = row['Gender']
    quality = [r.strip() for r in row['Roles/Titles'].split(';')]
    byear = int(row['Birth year'])
    dyear = int(row['Death year'])
    url = row['EMLO URL'] #.split('/')[-1]
    emlo_id = row['EMLO Person ID']
    uuid = row['UUID']
    name_publisher = name_publisher
    bdd = biodes.create_biodes_document(naam=name,
                                namen_en=variants,
                                geboortedatum_tekst=f'01-01-{byear}',
                                sterfdatum_tekst=f'31-12-{dyear}',
                                geslacht=gender_d.get(gender) or 2,
                                beroep=quality, 
                                name_publisher=name_publisher,
                                url_publisher='http://emlo.bodleian.ox.ac.uk/profile/person',
                                url_biography=url)
    return bdd
    
```

```python
import tarfile #import Tarfile
def compress_output_files(outputdir='.', outputfolder='out'):
    outputfile = os.path.join(outputdir, outputfolder)
    tar = tarfile.open(f"{outputfile}.tar.gz", "w:gz")
    for name in os.listdir(outputfile):
        tar.add(outputfile + "/" + name, arcname=name)
    tar.close()
```
